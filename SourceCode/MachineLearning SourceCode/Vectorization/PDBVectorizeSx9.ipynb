```python
# setting up function for building the Y vector from with the goddamnit library DSSP

from Bio.PDB.DSSP import DSSP
#dssp = DSSP(MODEL, './'+filename[1:3]+'/'+filename+'.cif', dssp='mkdssp')
#a_key=list(dssp.keys())[1]
#dssp[a_key]

def BuildVectorY(id1,id2):
    dsspTuple=dssp[(id1,id2)]
    yvec=None
    if dsspTuple[2]=='H' or dsspTuple[2]=='G' or dsspTuple[2]=='I':
        yvec=np.array([[1,0,0,0]])
    elif dsspTuple[2]=='B' or dsspTuple[2]=='E':
        yvec=np.array([[0,1,0,0]])
    elif dsspTuple[2]=='T' or dsspTuple[2]=='S':
        yvec=np.array([[0,0,1,0]])
    else:
        yvec=np.array([[0,0,0,1]])
    phipsi=np.array([[dsspTuple[4],dsspTuple[5]]])
    yvec=np.append(yvec,phipsi, axis=1)
    return yvec

def BuildVectorX(aminoacid,arr):
    obj=aapo[tri2sing(aminoacid)]
    resarr=np.array([[obj['averageMass'],obj['hydrogenBond'],obj['hydrophilicity'],obj['hydrophobicity'],obj['netChargeIndexOfSideChains'],obj['polarity'],obj['polarizability'],obj['solventAccessibleSurfaceArea'],obj['volumeOfSideChains']]])
    #return np.append(arr, resarr, axis=1)           
    return resarr
    
def tri2sing(aminoacid):
    if aminoacid=='GLY':
        return 'G'
    elif aminoacid=='ALA':
        return 'A'
    elif aminoacid=='LEU':
        return 'L'
    elif aminoacid=='MET':
        return 'M'
    elif aminoacid=='PHE':
        return 'F'
    elif aminoacid=='TRP':
        return 'W'
    elif aminoacid=='LYS':
        return 'K'
    elif aminoacid=='GLN':
        return 'Q'
    elif aminoacid=='GLU':
        return 'E'
    elif aminoacid=='SER':
        return 'S'
    elif aminoacid=='PRO':
        return 'P'
    elif aminoacid=='VAL':
        return 'V'
    elif aminoacid=='ILE':
        return 'I'
    elif aminoacid=='CYS':
        return 'C'
    elif aminoacid=='TYR':
        return 'Y'
    elif aminoacid=='HIS':
        return 'H'
    elif aminoacid=='ARG':
        return 'R'
    elif aminoacid=='ASN':
        return 'N'
    elif aminoacid=='ASP':
        return 'D'
    elif aminoacid=='THR':
        return 'T'
```
